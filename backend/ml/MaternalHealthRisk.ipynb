```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, mean_absolute_error
import xgboost as xgb
```

```python
df = pd.read_csv('/content/drive/MyDrive/kaggle/Maternal Health Risk Data Set.csv')
```

```python
df.head()
```

```python
df.RiskLevel.unique()
```

```python
df.shape
```

```python
df.dropna()
df.shape
```

```python
df['RiskLevel'] = df['RiskLevel'].apply(lambda x: 0 if x == 'low risk' else 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('RiskLevel', axis=1), df['RiskLevel'], test_size=0.2, random_state=1)
```

```python
model = RandomForestClassifier(n_estimators=100, random_state=1)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(mean_absolute_error(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
model_boost = xgb.XGBClassifier(n_estimators=500, random_state=1)
model_boost.fit(X_train, y_train)
```

```python
y_pred = model_boost.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
model_boost.save_model('maternal_health_model.bin')
```
